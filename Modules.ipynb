# Python Modules Complete Advanced Tutorial

# What is a Module?

```python
# Definition:
# A module is a file containing Python code.
# Modules help organize code into separate files.

# Example of importing module
import math

# Use module function
print(math.sqrt(25))
```

#  Why Modules are Important

```python
# Modules make code reusable and oraganized 

# without modules:
# Large  programs become difficult to manage

# with modules:
# code becomes clean and reusable

print("Modules improve code management")
```

# Importing a Module

```python
# import keyword loads module

import math

# use math module
print(math.pi)
```

# Using math Module

```python
# math module provides mathematical functions

import math

# Square root
print(math.sqrt(16))

# Power
print(
    math.pow(2, 3))

# Factorial 
print(math.factorial(5))
```

# Using random Module

```python
# random module generates random  values

import random

# Random number
print(random.randint(1, 10))

# Random choice
print(random.choice(["apple","banana","mango"]))
```

# Using Datetime Module 

```python
# datetime module works with date and time

import datetime

# Current date and time
now = datetime.datetime.now()

print(now)
```

# Import Specific Function

```python
# Import only required function

from math import sqrt

# Use sqrt directly
print(sqrt(36))
```

# Import with Alias

```python
# Alis gives short name to module

import math as m

# Use alias name
print(m.pi)
```

# Custom Module

# Using Statistics Module

```python
# Statistics module performs statistical operations

import statistics

data = [10,20,30,40]

# Mean value
print(statistics.mean(data))
```

# Using time Module
